# Dataset

```python
batch_size = 1

train_loader = torch.utils.data.DataLoader(
  torchvision.datasets.MNIST('.', train=True, download=True,
                             transform=torchvision.transforms.Compose([
                               torchvision.transforms.ToTensor(),
                               torchvision.transforms.Normalize(
                                 (0.1307,), (0.3081,))
                             ])),
  batch_size=batch_size, shuffle=True)

test_loader = torch.utils.data.DataLoader(
  torchvision.datasets.MNIST('.', train=False, download=True,
                             transform=torchvision.transforms.Compose([
                               torchvision.transforms.ToTensor(),
                               torchvision.transforms.Normalize(
                                 (0.1307,), (0.3081,))
                             ])),
  batch_size=batch_size, shuffle=True)
```

# Model Parameters

```python
layers = 2
heads = 14
dmodel = 28*28
dk = dv = int(dmodel / heads)

print(f'dmodel: {dmodel}')
print(f'dk: {dk}')
```

```python
device = 'mps' if torch.has_mps else 'gpu' if torch.cuda.is_available else 'cpu'

transformer = Transformer.Transformer(layers, heads, dk, dv, output_size=10)

optimiser = torch.optim.Adam(transformer.parameters(), lr=1e-6)
criterion = torch.nn.CrossEntropyLoss()
```

```python
# transformer = Transformer.Transformer(layers, heads, dk, dv, output_size=10)

# optimiser = torch.optim.Adam(transformer.parameters(), lr=1e-8)
# criterion = torch.nn.CrossEntropyLoss()

transformer.train()

for epoch in range(10):
    train_data = enumerate(train_loader)

    # total_loss = 0
    total_correct = 0
    for i in range(0, len(train_loader)):
        total_loss = 0
        optimiser.zero_grad()

        # Forward pass
        train_input, train_label = next(train_data)[1]
        train_input = train_input.flatten().unsqueeze(0)
        temp = np.zeros((10))
        temp[train_label[0]] = 1
        train_label = torch.Tensor(temp).float()

        prediction = transformer(train_input)

        prediction = prediction.squeeze(0)

        loss = criterion(prediction, train_label)
        total_loss += loss.item()

        # Back-Propagation
        loss.backward()
        optimiser.step()

        correct = prediction.argmax() == train_label.argmax()
        if correct:
            total_correct += 1

        print(f'Iter: {i}, Train Loss: {total_loss}, Correct: {correct}')

    print(f'Epoch: {epoch}, Total Correct: {total_correct}/{len(train_loader)}')
```

```python
transformer.eval()

test_data = enumerate(test_loader)

# total_loss = 0
total_correct = 0
for i in range(0, len(test_loader)):
    total_loss = 0

    # Forward pass
    test_input, test_label = next(test_data)[1]
    test_input = test_input.flatten().unsqueeze(0)
    temp = np.zeros((10))
    temp[test_label[0]] = 1
    test_label = torch.Tensor(temp).float()

    prediction = transformer(test_input)

    prediction = prediction.squeeze(0)

    correct = prediction.argmax() == test_label.argmax()
    if correct:
        total_correct += 1

    print(f'Iter: {i}, Correct: {correct}')

print(f'Epoch: {epoch}, Total Correct: {total_correct}/{len(test_loader)}')
```
